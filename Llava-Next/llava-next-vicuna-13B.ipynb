```python
from transformers import LlavaNextProcessor, LlavaNextForConditionalGeneration
import torch
from PIL import Image
import requests
from transformers import LlavaNextForConditionalGeneration, BitsAndBytesConfig
```

```python

# specify how to quantize the model
kwargs = {"device_map": "auto"}


kwargs['quantization_config'] = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float16,
)
processor = LlavaNextProcessor.from_pretrained("llava-hf/llava-v1.6-vicuna-13b-hf", token = "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")
```

```python
model = LlavaNextForConditionalGeneration.from_pretrained("llava-hf/llava-v1.6-vicuna-13b-hf", low_cpu_mem_usage=True, **kwargs)

```

```python
torch.cuda.empty_cache()
```

```python
from PIL import Image
import os
image_folder_path = r'/content/stuff'
# List all files in the directory
image_files = [f for f in os.listdir(image_folder_path) if f.endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))]

# Iterate through image files
for image_file in image_files:
    image_path = os.path.join(image_folder_path, image_file)
    image = Image.open(image_path)
    # Display or process the image
    image.show()  # This will display the image
    # Perform other operations on the image
    # For example, convert to grayscale:
    # gray_image = image.convert('L')
```

```python
print(image_files,'\n')
```

```python

folder_path = r'/content/stuff'
for i in range(len(image_files)):
  print('\n')
  print(i)
  print(image_files[i])
  image_path = os.path.join(folder_path, image_files[i])
  image = Image.open(image_path)
  question = """Read the given invoice and tell me the price information that you gathered from it."""
  prompt = f"""A chat between a curious human and an artificial intelligence assistant.
  The assistant gives helpful, detailed, and polite answers to the human's questions.
  USER: <image>\{question} ASSISTANT:"""

  inputs = processor(prompt, image, return_tensors="pt").to("cuda:0")

  # autoregressively complete prompt
  output = model.generate(**inputs, max_new_tokens=500, do_sample = True,temperature = 0.001)

  print(processor.decode(output[0], skip_special_tokens=True))
```
